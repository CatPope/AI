```python
from pykrx import stock
from pykrx import bond
#pykrx 모듈 import
```

```python
market = stock.get_market_ohlcv("20240101", "20241231", "005930")
market
#market이라는 변수에 2024년 1월 1일부터 2024년 12월 31일까지의 삼성전자 주식 변동을 수집
```

```python
from io import StringIO
csv_buffer = StringIO()
market.to_csv(csv_buffer)

csv_data = csv_buffer.getvalue()
print(csv_data)
#market 변수에 할당된 값을 String 형태로 변환
```

```python
import os
import openai

openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {
      "role": "system",
      "content": "다음은 삼성전자의 날짜별 주가 데이터야. 이 데이터를 분석해서 보고해줘."
    },
    {
      "role": "user",
      "content": csv_data
    }
  ],
  temperature=0,
  max_tokens=1024,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
