```python
import openai
from pykrx import stock

# 시장 데이터 가져오기
df = stock.get_market_ohlcv("20231201", "20240120", "005930")

# 데이터를 AI가 이해할 수 있는 형태로 요약 (문자열로 변환)
data = df.head(50).to_string()

# OpenAI API 요청 준비
res = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "다음 주식 데이터를 분석해주세요."},  
        {"role": "user", "content": data}
    ],
    temperature=0.5,
    max_tokens=1024,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0
)

# 분석 결과 출력
print(res.choices[0].message.content)
```
