## pre-trained model ê³¼ tokenizerë¥¼ pipeline ì— ì§€ì •í•˜ì—¬ ì‚¬ìš©

ì˜ˆë¥¼ ë“¤ì–´ "nlptown/bert-base-multilingual-uncased-sentiment" ëª¨ë¸ì€ ì˜ì–´, í•œêµ­ì–´, í”„ë‘ìŠ¤ì–´, ë„¤ëœë€ë“œì–´, ë…ì¼ì–´, ì´íƒˆë¦¬ì•„ì–´ ë° ìŠ¤í˜ì¸ì–´ë¡œ ëœ í…ìŠ¤íŠ¸ë¥¼ ì²˜ë¦¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì‚¬ì „ í›ˆë ¨ëœ ëª¨ë¸ì„ ì €ì¥í•œ ë¡œì»¬ í´ë”ë¡œ í•´ë‹¹ ì´ë¦„ì„ ë°”ê¿€ ìˆ˜ë„ ìˆìŠµë‹ˆë‹¤(ì•„ë˜ ì°¸ì¡°).  ëª¨ë¸ ê°œì²´ ë° ì—°ê²°ëœ í† í¬ë‚˜ì´ì €ë¥¼ ì „ë‹¬í•  ìˆ˜ë„ ìˆìŠµë‹ˆë‹¤.

ì´ë¥¼ ìœ„í•´ ë‘ ê°œì˜ í´ë˜ìŠ¤ê°€ í•„ìš”í•©ë‹ˆë‹¤. 

ì²« ë²ˆì§¸ëŠ” ìš°ë¦¬ê°€ ì„ íƒí•œ ëª¨ë¸ê³¼ ê´€ë ¨ëœ í† í¬ë‚˜ì´ì €ë¥¼ ë‹¤ìš´ë¡œë“œí•˜ê³  ì¸ìŠ¤í„´ìŠ¤í™”í•˜ëŠ” ë° ì‚¬ìš©í•  `AutoTokenizer`ì…ë‹ˆë‹¤. 

ë‘ ë²ˆì§¸ëŠ” 'AutoModelForSequenceClassification'(TensorFlowë¥¼ ì‚¬ìš©í•˜ëŠ” ê²½ìš° `TFAutoModelForSequenceClassification`)ëŠ” ëª¨ë¸ ìì²´ë¥¼ ë‹¤ìš´ë¡œë“œí•˜ëŠ” ë° ì‚¬ìš©í•  ê²ƒì…ë‹ˆë‹¤. 

ë‹¤ë¥¸ Task ì˜ ê²½ìš° ëª¨ë¸ì˜ í´ë˜ìŠ¤ê°€ ë³€ê²½ë©ë‹ˆë‹¤. [Task Summary](https://huggingface.co/transformers/task_summary.html) íŠœí† ë¦¬ì–¼ì—ëŠ” ì–´ë–¤ í´ë˜ìŠ¤ê°€ ì–´ë–¤ ì‘ì—…ì— ì‚¬ìš©ë˜ëŠ”ì§€ ìš”ì•½ë˜ì–´ ìˆìŠµë‹ˆë‹¤.

````python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline
````

ì´ì œ ìš°ë¦¬ê°€ ì°¾ì€ ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì €ë¥¼ ë‹¤ìš´ë¡œë“œí•˜ë ¤ë©´ `AutoModelForSequenceClassification.from_pretrained` ë©”ì„œë“œë¥¼ ì‚¬ìš©í•´ì•¼ í•©ë‹ˆë‹¤(`model_name`ì„ model hubì˜ ë‹¤ë¥¸ ëª¨ë¸ë¡œ ììœ ë¡­ê²Œ ëŒ€ì²´).

````python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
# This model only exists in PyTorch, so we use the `from_pt` flag to import that model in TensorFlow.
model = AutoModelForSequenceClassification.from_pretrained(model_name)

tokenizer = AutoTokenizer.from_pretrained(model_name)

classifier = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
````

````python
classifier('I love you'), classifier('ë‚œ ë„ ì‚¬ë‘í•´')
````

ìš°ë¦¬ê°€ ê°€ì§„ ë°ì´í„°ì™€ ìœ ì‚¬í•œ ë°ì´í„°ë¡œ ì‚¬ì „ í›ˆë ¨ëœ ëª¨ë¸ì„ ì°¾ì§€ ëª»í•œ ê²½ìš° ìš°ë¦¬ê°€ ê°€ì§„ ë°ì´í„°ì— ëŒ€í•´ ì‚¬ì „ í›ˆë ¨ëœ ëª¨ë¸ì„ ë¯¸ì„¸ ì¡°ì •í•´ì•¼ í•©ë‹ˆë‹¤. ì´ë¥¼ ìœ„í•´ [ì˜ˆì œ ìŠ¤í¬ë¦½íŠ¸](https://huggingface.co/transformers/examples.html)ë¥¼ ì œê³µí•©ë‹ˆë‹¤. ì™„ë£Œë˜ë©´ [ì´ íŠœí† ë¦¬ì–¼](https://huggingface.co/transformers/model_sharing.html)ì„ ì‚¬ìš©í•˜ì—¬ í—ˆë¸Œì—ì„œ ë¯¸ì„¸ ì¡°ì •ëœ ëª¨ë¸ì„ ì»¤ë®¤ë‹ˆí‹°ì™€ ê³µìœ í•˜ëŠ” ê²ƒì„ ìŠì§€ ë§ˆì‹­ì‹œì˜¤.

## Under the hood: pretrained models

ì´ì œ ì´ëŸ¬í•œ íŒŒì´í”„ë¼ì¸ì„ ì‚¬ìš©í•  ë•Œ ë‚´ë¶€ì—ì„œ ì–´ë–¤ ì¼ì´ ë°œìƒí•˜ëŠ”ì§€ ì‚´í´ë³´ê² ìŠµë‹ˆë‹¤.  

ìš°ë¦¬ê°€ ë³´ì•˜ë“¯ì´ ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì €ëŠ” `from_pretrained` ë©”ì†Œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ìƒì„±ë©ë‹ˆë‹¤.

````python
model_name = "distilbert-base-uncased-finetuned-sst-2-english"

pt_model = AutoModelForSequenceClassification.from_pretrained(model_name)

tokenizer = AutoTokenizer.from_pretrained(model_name)
````

### í† í¬ë‚˜ì´ì € ì‚¬ìš©

ìš°ë¦¬ëŠ” í† í¬ë‚˜ì´ì €ê°€ í…ìŠ¤íŠ¸ì˜ ì „ì²˜ë¦¬ë¥¼ ë‹´ë‹¹í•œë‹¤ê³  ì–¸ê¸‰í–ˆìŠµë‹ˆë‹¤. ë¨¼ì € ì£¼ì–´ì§„ í…ìŠ¤íŠ¸ë¥¼ ì¼ë°˜ì ìœ¼ë¡œ *token*ì´ë¼ê³  í•˜ëŠ” ë‹¨ì–´(ë˜ëŠ” ë‹¨ì–´ì˜ ì¼ë¶€, êµ¬ë‘ì  ê¸°í˜¸ ë“±)ë¡œ ë¶„í• í•©ë‹ˆë‹¤. í•´ë‹¹ í”„ë¡œì„¸ìŠ¤ë¥¼ ì œì–´í•  ìˆ˜ ìˆëŠ” ì—¬ëŸ¬ ê·œì¹™ì´ ìˆìŠµë‹ˆë‹¤([í† í°ë‚˜ì´ì € ìš”ì•½](https://huggingface.co/transformers/tokenizer_summary.html)ì—ì„œ ìì„¸íˆ ì•Œì•„ë³¼ ìˆ˜ ìˆìŒ). ì´ê²ƒì´ ëª¨ë¸ì´ ì‚¬ì „ í›ˆë ¨ë˜ì—ˆì„ ë•Œì™€ ë™ì¼í•œ ê·œì¹™ì„ ì‚¬ìš©í•˜ë„ë¡ í•˜ê¸° ìœ„í•´ ëª¨ë¸ì˜ ì´ë¦„ì„ ì‚¬ìš©í•˜ì—¬ í† í¬ë‚˜ì´ì €ë¥¼ ì¸ìŠ¤í„´ìŠ¤í™”í•´ì•¼ í•˜ëŠ” ì´ìœ ì…ë‹ˆë‹¤.

ë‘ ë²ˆì§¸ ë‹¨ê³„ëŠ” í•´ë‹¹ *token*ì„ ìˆ«ìë¡œ ë³€í™˜í•˜ì—¬ í…ì„œë¥¼ ë§Œë“¤ê³  ëª¨ë¸ì— ê³µê¸‰í•  ìˆ˜ ìˆë„ë¡ í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤. ì´ë¥¼ ìœ„í•´ í† í¬ë‚˜ì´ì €ëŠ” *vocab*ì„ ê°€ì§€ê³  ìˆëŠ”ë°, ì´ëŠ” ëª¨ë¸ì´ ì‚¬ì „ í›ˆë ¨ë˜ì—ˆì„ ë•Œì™€ ë™ì¼í•œ *vocab*ì„ ì‚¬ìš©í•´ì•¼ í•˜ê¸° ë•Œë¬¸ì— `from_pretrained` ë©”ì„œë“œë¡œ ì¸ìŠ¤í„´ìŠ¤í™”í•  ë•Œ ë‹¤ìš´ë¡œë“œí•˜ëŠ” ë¶€ë¶„ì…ë‹ˆë‹¤.

ì£¼ì–´ì§„ í…ìŠ¤íŠ¸ì— ì´ ë‹¨ê³„ë¥¼ ì ìš©í•˜ë ¤ë©´ í† í¬ë‚˜ì´ì €ì— ì…ë ¥í•˜ë©´ ë©ë‹ˆë‹¤.

````python
inputs = tokenizer("We are very happy to show you the ğŸ¤— Transformers library.")
````

ì´ê²ƒì€ integer listë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤. ì—¬ê¸°ì—ëŠ” ì•ì„œ ì–¸ê¸‰í•œ [í† í°ì˜ Id](https://huggingface.co/transformers/glossary.html#input-ids)ë¿ ì•„ë‹ˆë¼ ëª¨ë¸ì— ìœ ìš©í•œ ì¶”ê°€ ì¸ìˆ˜ë„ í¬í•¨ë©ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´ ëª¨ë¸ì´ ì‹œí€€ìŠ¤ë¥¼ ë” ì˜ ì´í•´í•  ìˆ˜ë„ë¡ í•˜ê¸° ìœ„í•œ [attention mask](https://huggingface.co/transformers/glossary.html#attention-mask)ë„ ìˆìŠµë‹ˆë‹¤.

````python
print(inputs)
````

sequence listë¥¼ í† í¬ë‚˜ì´ì €ì— ì§ì ‘ ì „ë‹¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ëª¨ë¸ì„ í†µí•´ ì¼ê´„ ì²˜ë¦¬ í•˜ë ¤ë©´ ëª¨ë‘ ê°™ì€ ê¸¸ì´ë¡œ paddingí•˜ê³  ëª¨ë¸ì´ í—ˆìš©í•  ìˆ˜ ìˆëŠ” ìµœëŒ€ ê¸¸ì´ë¡œ truncate í•œ í…ì„œë¥¼ ë§Œë“¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ ëª¨ë“  ê²ƒì„ í† í¬ë‚˜ì´ì €ì— ì§€ì •í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
pt_batch = tokenizer(
        ["We are very happy to show you the ğŸ¤— Transformers library.", 
         "We hope you don't hate it."],
        padding=True,
        truncation=True,
        max_length=512,
        return_tensors="pt"
)
````

íŒ¨ë”©ì€ ëª¨ë¸ì´ ë¯¸ë¦¬ í›ˆë ¨í•œ íŒ¨ë”© í† í°ê³¼ í•¨ê»˜ ëª¨ë¸ì´ ì˜ˆìƒí•˜ëŠ” ë©´(ì´ ê²½ìš° ì˜¤ë¥¸ìª½)ì— ìë™ìœ¼ë¡œ ì ìš©ë©ë‹ˆë‹¤. attention maskëŠ” íŒ¨ë”©ì„ ê³ ë ¤í•˜ë„ë¡ ì¡°ì •ë˜ì—ˆìŠµë‹ˆë‹¤.

````python
for key, value in pt_batch.items():
    print(f"{key}: {value.numpy().tolist()}")
````

[ì—¬ê¸°](https://huggingface.co/transformers/preprocessing.html)ì—ì„œ í† í¬ë‚˜ì´ì €ì— ëŒ€í•´ ìì„¸íˆ ì•Œì•„ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

### ëª¨ë¸ ì‚¬ìš©

ì…ë ¥ì´ í† í¬ë‚˜ì´ì €ì— ì˜í•´ ì‚¬ì „ ì²˜ë¦¬ë˜ë©´ ëª¨ë¸ì— ì§ì ‘ ë³´ë‚¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì–¸ê¸‰í–ˆë“¯ì´ ëª¨ë¸ì— í•„ìš”í•œ ëª¨ë“  ê´€ë ¨ ì •ë³´ê°€ í¬í•¨ë©ë‹ˆë‹¤. TensorFlow ëª¨ë¸ì„ ì‚¬ìš©í•˜ëŠ” ê²½ìš° dictionary keyë¥¼ í…ì„œì— ì§ì ‘ ì „ë‹¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. PyTorch ëª¨ë¸ì˜ ê²½ìš° `**`ë¥¼ ì¶”ê°€í•˜ì—¬ dictionary ì••ì¶•ì„ í’€ì–´ì•¼ í•©ë‹ˆë‹¤.

````python
pt_outputs = pt_model(**pt_batch)
````

ğŸ¤— Transformersì—ì„œ ëª¨ë“  ì¶œë ¥ì€ ë‹¤ë¥¸ ë©”íƒ€ë°ì´í„°ì™€ í•¨ê»˜ ëª¨ë¸ì˜ final activationì„ í¬í•¨í•˜ëŠ” objectì…ë‹ˆë‹¤. ì´ëŸ¬í•œ objectëŠ” [ì—¬ê¸°](https://huggingface.co/transformers/main_classes/output.html)ì— ëŒ€í•´ ìì„¸íˆ ì„¤ëª…ë˜ì–´ ìˆìŠµë‹ˆë‹¤. ì§€ê¸ˆì€ ì¶œë ¥ì„ ì§ì ‘ ê²€ì‚¬í•´ ë³´ê² ìŠµë‹ˆë‹¤.

````python
print(pt_outputs)
print()
print(pt_outputs[0])
````

ì¶œë ¥ ê°ì²´ì— ì–´ë–»ê²Œ `logits` ì†ì„±ì´ ìˆëŠ”ì§€ ì£¼ëª©í•˜ì„¸ìš”. ì´ë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ì˜ ìµœì¢… activationì— ì•¡ì„¸ìŠ¤í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

**ì°¸ê³ :** ëª¨ë“  ğŸ¤— íŠ¸ëœìŠ¤í¬ë¨¸ ëª¨ë¸(PyTorch ë˜ëŠ” TensorFlow)ì€ ìµœì¢… í™œì„±í™” í•¨ìˆ˜ê°€ ì¢…ì¢… ì†ì‹¤ê³¼ ìœµí•©ë˜ê¸° ë•Œë¬¸ì— ìµœì¢… í™œì„±í™” í•¨ìˆ˜(SoftMaxì™€ ê°™ì€) ì „ì˜ logit ê°’ì„ ë°˜í™˜í•©ë‹ˆë‹¤.

ì˜ˆì¸¡ì„ ì–»ê¸° ìœ„í•´ SoftMax activationì„ ì ìš©í•´ ë³´ê² ìŠµë‹ˆë‹¤.

````python
from torch import nn
pt_predictions = nn.functional.softmax(pt_outputs.logits, dim=-1)
print(pt_predictions)
````

ëª¨ë¸ì— ì…ë ¥ ì™¸ì— ë ˆì´ë¸”ì„ ì œê³µí•˜ë©´ ëª¨ë¸ ì¶œë ¥ ê°ì²´ì—ë„ 'loss' ì†ì„±ì´ í¬í•¨ë©ë‹ˆë‹¤.

````python
import torch
pt_outputs = pt_model(**pt_batch, labels = torch.tensor([1, 0]))
print(pt_outputs)
print()
print(pt_outputs[0])
print(pt_outputs[1])
````

ëª¨ë¸ì€ í‘œì¤€ [torch.nn.Module](https://pytorch.org/docs/stable/nn.html#torch.nn.Module) ë˜ëŠ” [tf.keras.Model](https://www.tensorflow. org/api_docs/python/tf/keras/Model)ì„ ì‚¬ìš©í•˜ì—¬ ì¼ë°˜ì ì¸ êµìœ¡ ë£¨í”„ì—ì„œ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. 

ğŸ¤—TransformersëŠ” ë˜í•œ í›ˆë ¨ì„ ë•ê¸° ìœ„í•´ `Trainer`(ë˜ëŠ” TensorFlowë¥¼ ì‚¬ìš©í•˜ëŠ” ê²½ìš° `TFTrainer`) í´ë˜ìŠ¤ë¥¼ ì œê³µí•©ë‹ˆë‹¤(ë¶„ì‚° í›ˆë ¨, mixed precision ë“± ì²˜ë¦¬). ìì„¸í•œ ë‚´ìš©ì€ [êµìœ¡ íŠœí† ë¦¬ì–¼](https://huggingface.co/transformers/training.html)ì„ ì°¸ì¡°í•˜ì„¸ìš”.

ëª¨ë¸ì´ ë¯¸ì„¸ ì¡°ì •ë˜ë©´ ë‹¤ìŒê³¼ ê°™ì€ ë°©ì‹ìœ¼ë¡œ ëª¨ë¸ì„ ì €ì¥í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```
tokenizer.save_pretrained(save_directory)
```

ê·¸ëŸ° ë‹¤ìŒ ëª¨ë¸ ì´ë¦„ ëŒ€ì‹  ë””ë ‰í† ë¦¬ ì´ë¦„ì„ ì „ë‹¬í•˜ì—¬ `AutoModel.from_pretrained` ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì´ ëª¨ë¸ì„ ë‹¤ì‹œ ë¡œë“œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. 

ğŸ¤— Transformersì˜ ë©‹ì§„ ê¸°ëŠ¥ ì¤‘ í•˜ë‚˜ëŠ” PyTorchì™€ TensorFlow ê°„ì— ì‰½ê²Œ ì „í™˜í•  ìˆ˜ ìˆë‹¤ëŠ” ê²ƒì…ë‹ˆë‹¤. ì´ì „ì— ì €ì¥í•œ ëª¨ë“  ëª¨ë¸ì€ PyTorch ë˜ëŠ” TensorFlowì—ì„œ ë‹¤ì‹œ ë¡œë“œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. TensorFlow ëª¨ë¸ì— ì €ì¥ëœ PyTorch ëª¨ë¸ì„ ë¡œë“œí•˜ëŠ” ê²½ìš° ë‹¤ìŒê³¼ ê°™ì´ `TFAutoModel.from_pretrained`ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.

```

```
