### Flickr8k Dataset 전처리

- Flickr8k.token.txt 에서 caption 을 읽어 whitespace, punctuation 등 제거  

```
line => 
1000268201_693b08cb0e.jpg#0	A child in a pink dress is climbing up a set of stairs in an entry way .
line.split() =>
['1000268201_693b08cb0e.jpg#1', 'A', 'girl', 'going', 'into', 'a', 'wooden', 'building', '.']
```
```
ascii : ! 33, " 34, # 35, $ 36, % 37, & 38, '  39	
```

````python
print(string.punctuation)
print(str.maketrans('', '', string.punctuation))
````

````python
filename = root_captioning + "/Flickr8k_text/Flickr8k.token.txt"

fp = open(filename, 'r')

null_punct = str.maketrans('', '', string.punctuation)
lookup = dict()

max_length = 0

for line in fp.read().split('\n'):      
    tokens = line.split()
    if len(line) >= 2:
        id = tokens[0].split('.')[0]       #1000268201_693b08cb0e
        desc = tokens[1:]       #['A', 'girl', 'going', 'into', 'a', 'wooden', 'building', '.']
       # Cleanup description
        desc = [word.lower() for word in desc]
        desc = [w.translate(null_punct) for w in desc]
        desc = [word for word in desc if len(word)>1]    # a 같은 한글자 단어 제외
        desc = [word for word in desc if word.isalpha()]  # 알파벳 문자외는 제외
        max_length = max(max_length, len(desc))        # 가장 긴 description 길이 (단어수)

        if id not in lookup:
            lookup[id] = list()
        lookup[id].append(' '.join(desc))

print(lookup['1000268201_693b08cb0e'])
````

**lookup table 의 description 을 이용한 단어 사전 작성**

````python
vocaburary = set()
for key in lookup:
    [vocaburary.update(d.split()) for d in lookup[key]]

print(len(vocaburary))          # 전체 unique 단어 갯수
print(max_length)                # 가장 긴 caption 길이 (in words)
````

### Transfer Learning - pre-trained Neural Network 

-  pre-trained model load and extract features from the loaded photos

- InceptionV3 의 input shape 은 (299, 299, 3)

````python
encode_model = InceptionV3(weights='imagenet')
encode_model = Model(encode_model.input, encode_model.layers[-2].output)
WIDTH = 299
HEIGHT = 299
OUTPUT_DIM = 2048
preprocess_input = tensorflow.keras.applications.inception_v3.preprocess_input
````

````python
encode_model.summary()
````

### Training / Test Set 생성

- image 를 encode 하여 training set 생성  
- encoding 한 image 를 RNN 의 입력으로 제공하여 captioning  
    - pre-trained model 의 standard size 로 모든 image 를 resize (보간법은 ANTIALIAS)
    - PIL image 를 numpy array 로 변환
    - pre-trained model 이 요구하는 pre-processing 수행  
    

- 모든 JPG file 에 대해 loop  
- 실제 training 단계에서 모든 image 에서 feature 를 뽑아내는 작업은  시간이 오래 걸리므로 train 및 test image 의 feature encoding 결과물을 pickle file 로 저장 하여 실제 train 시 사용

````python
def encodeImage(img):
    img = img.resize((WIDTH, HEIGHT), Image.ANTIALIAS)
    x = tensorflow.keras.preprocessing.image.img_to_array(img)
    # Expand to 2D array
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    x = encode_model.predict(x) # Get the encoding vector for the image
    # Shape to correct form to be accepted by LSTM captioning network.
    x = np.reshape(x, OUTPUT_DIM )    #2048
    return x
````

````python
train_path = root_captioning + f'/train{OUTPUT_DIM}.pkl'

if not os.path.exists(train_path):
    start = time()
    encoding_train = {}

    for id in tqdm(train_img):
        image_path = os.path.join(img_directory, id)
        img = load_img(image_path, target_size=(HEIGHT, WIDTH))
        encoding_train[id] = encodeImage(img)

    with open(train_path, "wb") as fp:
        pickle.dump(encoding_train, fp)
    print(f"\nTotal 작업시간: {time()-start}")
else:
    with open(train_path, "rb") as fp:
        encoding_train = pickle.load(fp)
````

````python
test_path = root_captioning + f'/test{OUTPUT_DIM}.pkl'

if not os.path.exists(test_path):
    start = time()
    encoding_test = {}

    for id in tqdm(test_img):
        image_path = os.path.join(img_directory, id)
        img = load_img(image_path, target_size=(HEIGHT, WIDTH))
        encoding_test[id] = encodeImage(img)

    with open(test_path, "wb") as fp:
        pickle.dump(encoding_test, fp)
    print(f"\nTotal 작업시간: {time()-start}")
else:
    with open(test_path, "rb") as fp:
        encoding_test = pickle.load(fp)
````

### training set 에 image 와 함께 사용될 caption 분리

````python
all_train_captions = []

for key, val in train_descriptions.items():
    for cap in val:
        all_train_captions.append(cap)
        
print(len(all_train_captions))
all_train_captions[:3]
````

all_train_captions 에 출현 빈도가 10 회 미만인 단어는 삭제하여 전체 vocaburary size 를 줄인다.

````python
word_count_threshold = 10
word_counts = {}
nsents = 0

for sent in all_train_captions:
    nsents += 1
    for w in sent.split(' '):
        word_counts[w] = word_counts.get(w, 0) + 1

vocab = [w for w in word_counts if word_counts[w] >= word_count_threshold]
print('preprocessed words %d ==> %d' % (len(word_counts), len(vocab)))
vocab[:10]
````

### 최종 vocaburary 에 대해 index_to_word, word_to_index lookup table 작성
- index 는 1 부터 시작  

````python
index_to_word = {}
word_to_index = {}

ix = 1
for w in vocab:
    word_to_index[w] = ix
    index_to_word[ix] = w
    ix += 1
    
vocab_size = len(index_to_word) + 1 
vocab_size
````

### start / stop token(startseq, endseq) 가  모든 sentences 에 추가된 것을 maximum length of captions 에 반영

### Data Generator 생성

- Flikr8K data set 은 각 image 마다 5 개의 caption 으로 구성되어 있고, 모든 caption 의 각 timestep 마다 image 를 중복시켜 공급하기 때문에 전체 size 가 증가하여 memory 부족이 발생하지 않도록 generator 를 생성하여 data feed


- 다음 그림과 같은 sequence data 를 model 에 공급

<img src = https://raw.githubusercontent.com/jeffheaton/t81_558_deep_learning/master/images/caption-2.png width="700"/>

````python
def data_generator(descriptions, photos, wordtoidx, max_length, num_photos_per_batch):
    # x1 - Training data - 사진
    # x2 - 각 사진에 대한 caption 
    # y - x2 에 뒤따르는 label 부분 (word-by-word model)
    x1, x2, y = [], [], []
    n=0
    while True:
        for key, desc_list in descriptions.items():
            n+=1
            photo = photos[key+'.jpg']
           # 각 사진별로 5 개의 description 존재
            for desc in desc_list:
            # description 을 sequence 로 변환
                seq = [wordtoidx[word] for word in desc.split(' ') if word in wordtoidx]
            # sequence 의 각 timestep 별로 1 offset 만큼 shift 된 in, out sequence pair 생성
                for i in range(1, len(seq)):
                    in_seq, out_seq = seq[:i], seq[i]
                    in_seq = pad_sequences([in_seq], maxlen=max_length)[0]
                    out_seq = to_categorical([out_seq], num_classes=vocab_size)[0]
                    x1.append(photo)
                    x2.append(in_seq)
                    y.append(out_seq)
            # 1 batch generate       
            if n==num_photos_per_batch:
                yield ([np.array(x1), np.array(x2)], np.array(y))
                x1, x2, y = [], [], []
                n=0
````

### Loading Glove Embeddings

````python
embeddings_index = {}

with open(root_captioning + f'/glove.6B.{EMBEDDING_DIM}d.txt', encoding="utf8") as f:
    for line in tqdm(f):
        values = line.split()                 # 각 줄을 읽어와서 word_vector에 저장
        word = values[0]                    # 첫번째 값은 word
        coefs = np.asarray(values[1:], dtype='float32')  # 두번째 element 부터 마지막까지 100 개는 해당 단어의 임베딩 벡터의  값
        embeddings_index[word] = coefs
        
print(f'Found {len(embeddings_index)} word vectors.')
````

### word embedding matrix 작성

- embedding matrix 는 Glove 의 weight 를 copy   

- tokenizer 에서 작성된 word dictionary (word_to_index) 를 해당되는 단어의 pretrained embedding index 값으로 채움

- pretrained index 에 해당 단어 없으면 all zero 로 채움

````python
embedding_matrix = np.zeros((vocab_size, EMBEDDING_DIM))

for word, i in word_to_index.items(): 
    embedding_vector = embeddings_index.get(word)
    # Words not found in the embedding index will remain all zeros
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
````

embedding_matrix size 는 1652 (the size of the vocabulary) X  EMBEDDING_DIM

````python
embedding_matrix.shape
````

### Model 작성
- image feature extraction, sequence 

````python
# feature extractor model
inputs1 = Input(shape=(OUTPUT_DIM,))
fe1 = Dropout(0.5)(inputs1)
fe2 = Dense(256, activation='relu')(fe1)

# sequence model
inputs2 = Input(shape=(max_length,))
se1 = Embedding(vocab_size, EMBEDDING_DIM, mask_zero=True)(inputs2)
se2 = Dropout(0.5)(se1)
se3 = LSTM(256)(se2)

# Decoder model
decoder1= Add()([fe2, se3])
decoder2 = Dense(256, activation='relu')(decoder1)
outputs = Dense(vocab_size, activation='softmax')(decoder2)

caption_model = Model(inputs=[inputs1, inputs2], outputs=outputs)

caption_model.summary()
plot_model(caption_model, show_shapes=True)
````

### 작성한 model 의 embedding layer 에 embedding_matrix 값을 loading 하고 trainable=False 로 locking 함

````python
caption_model.layers[2].set_weights([embedding_matrix])
caption_model.layers[2].trainable = False

caption_model.compile(loss='categorical_crossentropy', optimizer='adam')
````

### Train the Neural Network

````python
number_pics_per_batch = 3
steps = len(train_descriptions)//number_pics_per_batch
steps
````

````python
NEW_TRAIN = True
EPOCHS = 10
model_path = root_captioning + '/caption-model.hdf5'
start = time()

if NEW_TRAIN:
    for i in tqdm(range(EPOCHS*2)):
        generator = data_generator(train_descriptions, encoding_train, word_to_index, max_length, number_pics_per_batch)
        caption_model.fit_generator(generator, epochs=1, steps_per_epoch=steps, verbose=1)

    caption_model.optimizer.lr = 1e-4
    number_pics_per_bath = 6
    steps = len(train_descriptions)//number_pics_per_batch  

    for i in range(EPOCHS):
        generator = data_generator(train_descriptions, encoding_train, word_to_index, max_length, number_pics_per_batch)
        caption_model.fit_generator(generator, epochs=1, steps_per_epoch=steps, verbose=1)  
        
    caption_model.save_weights(model_path)
    print(f"\Training 경과 시간: {time()-start}")
else:
    caption_model.load_weights(model_path)
````

### Caption 생성

- 두개의 입력 필요
    - photo  
    - caption - starting token 으로 시작.  
    
- neural network 의 output 은 caption 의 다음 word 이므로 end token 이 나오거나 caption 의 maximum length 에 도달할 때 까지 반복

````python
def generateCaption(photo):
    in_text = START
    for i in range(max_length):
        sequence = [word_to_index[w] for w in in_text.split() if w in word_to_index]
        sequence = pad_sequences([sequence], maxlen=max_length)
        yhat = caption_model.predict([photo, sequence], verbose=0)
        yhat = np.argmax(yhat)
        word = index_to_word[yhat]
        in_text += ' ' + word
        if word == STOP:
            break
    final = in_text.split()
    final = final[1:-1]     # START, STOP 제거
    final = ' '.join(final)
    return final
````
