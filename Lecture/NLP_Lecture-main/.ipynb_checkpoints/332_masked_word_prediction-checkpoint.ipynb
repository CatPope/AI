# 332. masked word prediction

```python
from transformers import BertTokenizerFast, BertForMaskedLM
import torch
import string
import re

tokenizer = BertTokenizerFast.from_pretrained('kykim/bert-kor-base')
model = BertForMaskedLM.from_pretrained('kykim/bert-kor-base').eval()
```

- BertForMaskedLM은 LM(Language Modeling) 헤드를 사용합니다.

- 일반적으로 LM 헤드는 은닉 상태의 입력 차원(BERT 기반의 경우 768)과 어휘 크기의 출력 차원을 갖는 선형 계층입니다. 따라서 BERT 모델의 숨겨진 상태 출력을 어휘의 특정 토큰에 매핑합니다. 

```python
text = '이거 [MASK] 재밌네?'

input_ids = torch.tensor([tokenizer.encode(text, add_special_tokens=True)])
input_ids
```

```python
torch.where(input_ids == 4)
```

[MASK] token의 index 를 구함.

```python
mask_idx = torch.where(input_ids == 4)[1].tolist()[0]
print(mask_idx)
```

```python
ignore_tokens = string.punctuation + '[PAD][UNK]<pad><unk> '

tokens = []
with torch.no_grad():
    prediction = model(input_ids)

prediction, prediction[0].shape
```

```python
import torch.nn.functional as F

out = F.softmax(prediction[0], dim=-1).argmax(dim=2)
out
```

```python
tokenizer.decode(5554), tokenizer.decode(14568), tokenizer.decode(2033)
```

```python
tokenizer.decode(out[0])
```

```python
top_clean = 3
pred_idx = prediction[0, mask_idx, :].topk(10).indices.tolist()
print(pred_idx)

for w in pred_idx:
    token = ''.join(tokenizer.decode(w).split())
    if token not in ignore_tokens:
        tokens.append(token.replace('##', ''))

for token in tokens[:top_clean]:
    print(re.sub('\[MASK\]', token, text))
```
