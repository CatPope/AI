## ê°ì • ë¶„ì„

```python
# ê°ì„± ë¶„ì„(sentiment analysis) íŒŒì´í”„ë¼ì¸ ìƒì„±
# ê¸°ë³¸ì ìœ¼ë¡œ ì˜ì–´ ë¬¸ì¥ì— ëŒ€í•´ ê¸ì •(positive) ë˜ëŠ” ë¶€ì •(negative) ê°ì •ì„ ë¶„ë¥˜í•˜ëŠ” ì‚¬ì „í•™ìŠµ ëª¨ë¸ì„ ì‚¬ìš©
classifier = pipeline('sentiment-analysis')
```

- ì´ ëª…ë ¹ì„ ì²˜ìŒ ì…ë ¥í•˜ë©´ ì‚¬ì „ í›ˆë ¨ëœ ëª¨ë¸ê³¼ í•´ë‹¹ í† í¬ë‚˜ì´ì €ê°€ ë‹¤ìš´ë¡œë“œë˜ì–´ ìºì‹œë©ë‹ˆë‹¤.  

- í† í¬ë‚˜ì´ì €ëŠ” ëª¨ë¸ì˜ í…ìŠ¤íŠ¸ë¥¼ ì‚¬ì „ ì²˜ë¦¬í•œ ë‹¤ìŒ ì˜ˆì¸¡ì„ ë‹´ë‹¹í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤.  

- íŒŒì´í”„ë¼ì¸ì€ ëª¨ë“  ê²ƒì„ í•¨ê»˜ ê·¸ë£¹í™”í•˜ê³  ì˜ˆì¸¡ì„ ì½ì„ ìˆ˜ ìˆë„ë¡ ì‚¬í›„ ì²˜ë¦¬í•©ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´:

```python
classifier('We are very happy to show you the ğŸ¤— Transformers library.')
```

ì‚¬ì „ ì²˜ë¦¬ëœ ë‹¤ìŒ ëª¨ë¸ì— *batch*ë¡œ ê³µê¸‰í•˜ë©´ ë‹¤ìŒê³¼ ê°™ì€ ì‚¬ì „ ëª©ë¡ì„ ë°˜í™˜í•©ë‹ˆë‹¤.

```python
results = classifier([
                      "We are very happy to show you the ğŸ¤— Transformers library.",
                       "We hope you don't hate it."
                        ])

for result in results:
    print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

## í•œêµ­ì–´ ê°ì •ë¶„ì„

- NSMC(Naver Sentiment Movie Corpus) ë¡œ ë¯¸ì„¸ ì¡°ì •ëœ BERT ë‹¤êµ­ì–´ basecase ëª¨ë¸ : https://huggingface.co/sangrimlee/bert-base-multilingual-cased-nsmc

```python
# í•œêµ­ì–´ ê°ì„± ë¶„ì„ íŒŒì´í”„ë¼ì¸ ìƒì„±
# 'sangrimlee/bert-base-multilingual-cased-nsmc' ëª¨ë¸ì€
# ë„¤ì´ë²„ ì˜í™” ë¦¬ë·° ë°ì´í„°ì…‹(NSMC)ìœ¼ë¡œ í•™ìŠµëœ ë‹¤êµ­ì–´ BERT ê¸°ë°˜ ê°ì„± ë¶„ë¥˜ ëª¨ë¸
classifier_ko = pipeline('sentiment-analysis',
                         model="sangrimlee/bert-base-multilingual-cased-nsmc")
```

- ìë™ ë³„ì  ë¶€ì—¬

```python
# ë‹¤êµ­ì–´ ê°ì„± ë¶„ì„ ëª¨ë¸ ì´ë¦„ ì§€ì •
# 'nlptown/bert-base-multilingual-uncased-sentiment'ëŠ” ì˜ì–´ë¿ë§Œ ì•„ë‹ˆë¼ í•œêµ­ì–´ ë“± ì—¬ëŸ¬ ì–¸ì–´ì˜ ê°ì„± ë¶„ì„ì´ ê°€ëŠ¥í•œ BERT ê¸°ë°˜ ëª¨ë¸
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"

# ê°ì„± ë¶„ì„ íŒŒì´í”„ë¼ì¸ ìƒì„± (í•œêµ­ì–´ ë¬¸ì¥ë„ ì§€ì›ë¨)
kor_classifier = pipeline('sentiment-analysis', model=model_name)

texts = ["ë‹¤ì‹œëŠ” ë³´ê³  ì‹¶ì§€ ì•Šì€ ì§œì¦ë‚˜ëŠ” ì˜í™”",
            "ì•„ì£¼ ì¬ë¯¸ìˆëŠ” ì˜í™”",
            "ì •ë§ ì¬ë¯¸ì—†ëŠ” ì˜í™”ì˜€ë‹¤",
            "ì´ ì˜í™” ë§í• ê±°ì•¼",
            "ì´ ì˜í™” ìµœê³ ",
            "ë³´í†µ ì˜í™”"]

results = kor_classifier(texts)

for i, result in enumerate(results):
    print(f"{texts[i]} --> ë³„ì : {result['label']}, with score: {round(result['score'], 4)}")
```

## Zero Shot Pipeline - ì²˜ìŒ ë³´ëŠ” ë¬¸ì¥ì˜ category ë¶„ë¥˜

```python
# ì œë¡œìƒ· ë¶„ë¥˜(zero-shot classification) íŒŒì´í”„ë¼ì¸ ìƒì„±
# ì‚¬ì „ ì •ì˜ëœ ë ˆì´ë¸”(label)ì— ëŒ€í•´ í•™ìŠµ ì—†ì´ ë¬¸ì¥ì„ ë¶„ë¥˜í•  ìˆ˜ ìˆìŒ
classifier = pipeline("zero-shot-classification")

# ì…ë ¥ ë¬¸ì¥ì„ ì£¼ì–´ì§„ í›„ë³´ ë ˆì´ë¸” ì¤‘ ì–´ë–¤ ë²”ì£¼ë¡œ ë¶„ë¥˜í• ì§€ ì˜ˆì¸¡
sequence = "This is a course about the Transformers library"
candidate_labels = ["education", "politics", "business"]

result = classifier(
    sequence     ,  # ë¶„ë¥˜í•  ë¬¸ì¥
    candidate_labels=candidate_labels  # í›„ë³´ ë ˆì´ë¸” ëª©ë¡
)

result
```

### Zero Shot Pipeline - ë‹¤êµ­ì–´ version

```python
classifier = pipeline("zero-shot-classification", model='joeddav/xlm-roberta-large-xnli')

sequence = "2025ë…„ ëŒ€í†µë ¹ ì„ ê±°ì— ëˆ„êµ¬ì—ê²Œ íˆ¬í‘œí•˜ê² ìŠµë‹ˆê¹Œ"
candidate_labels = ["ê±´ê°•", "ë¬¸í™”", "ì •ì¹˜"]
hypothesis_template = "ì´ í…ìŠ¤íŠ¸ëŠ” {}ì— ê´€í•œ ë‚´ìš© ì…ë‹ˆë‹¤."

result = classifier(sequence, candidate_labels, hypothesis_template=hypothesis_template)
result
```

## Text ìƒì„±

```python
# í…ìŠ¤íŠ¸ ìƒì„±(text generation) íŒŒì´í”„ë¼ì¸ ìƒì„±
# distilgpt2 ëª¨ë¸ì€ GPT-2ë¥¼ ê²½ëŸ‰í™”í•œ ë²„ì „ìœ¼ë¡œ, ë¹ ë¥´ê³  íš¨ìœ¨ì ì¸ í…ìŠ¤íŠ¸ ìƒì„±ì´ ê°€ëŠ¥í•¨
generator = pipeline("text-generation", model="distilgpt2")

# ì£¼ì–´ì§„ ë¬¸ì¥ìœ¼ë¡œë¶€í„° í…ìŠ¤íŠ¸ë¥¼ ì´ì–´ì„œ ìƒì„±
# - max_length: ì „ì²´ ìƒì„± ë¬¸ì¥ì˜ ìµœëŒ€ ê¸¸ì´
# - num_return_sequences: ìƒì„±í•  ë¬¸ì¥ ìˆ˜ (ì—¬ê¸°ì„  2ê°œ)
result = generator(
    "In this course, we will teach you how to",  # ì‹œì‘ ë¬¸ì¥
    max_length=30,
    num_return_sequences=2
)

print(result)
```

### ë§ˆìŠ¤í‚¹ëœ í…ìŠ¤íŠ¸ ì±„ìš°ê¸°

```python
# ë¹ˆì¹¸ ì±„ìš°ê¸° (Masked Language Modeling) íŒŒì´í”„ë¼ì¸ ìƒì„±
# BERT ê³„ì—´ ëª¨ë¸ì„ ì‚¬ìš©í•˜ì—¬ <mask> í† í°ì— ë“¤ì–´ê°ˆ ê°€ì¥ ì ì ˆí•œ ë‹¨ì–´ ì˜ˆì¸¡
unmasker = pipeline("fill-mask")

# ë¬¸ì¥ì—ì„œ <mask> ìœ„ì¹˜ì— ì˜¬ ìˆ˜ ìˆëŠ” ë‹¨ì–´ë¥¼ ì˜ˆì¸¡
# - top_k=2: ìƒìœ„ 2ê°œì˜ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ë°˜í™˜
result = unmasker("This course will teach you all about <mask> models.", top_k=2)

# ì˜ˆì¸¡ ê²°ê³¼ ì¶œë ¥
print(result)
```

### ê°œì²´ëª… ì¸ì‹

```python
# ê°œì²´ëª… ì¸ì‹(NER: Named Entity Recognition) íŒŒì´í”„ë¼ì¸ ìƒì„±
# grouped_entities=True: ê°™ì€ ê°œì²´ë¡œ ì¸ì‹ëœ ì—°ì†ëœ í† í°ë“¤ì„ í•˜ë‚˜ë¡œ ë¬¶ì–´ì„œ ë°˜í™˜
ner = pipeline("ner", grouped_entities=True)

# ì…ë ¥ ë¬¸ì¥ì—ì„œ ì‚¬ëŒ ì´ë¦„, ê¸°ê´€ëª…, ìœ„ì¹˜ ë“± ê³ ìœ  ëª…ì‚¬ë¥¼ ì¸ì‹
result = ner("My name is Sylvian and I work at Hugging Face in Brooklyn")

# ê²°ê³¼ ì¶œë ¥
print(result)
```

### ì§ˆì˜ ì‘ë‹µ

```python
# ì§ˆì˜ì‘ë‹µ(question answering) íŒŒì´í”„ë¼ì¸ ìƒì„±
# ì£¼ì–´ì§„ ë¬¸ë§¥(context)ì—ì„œ ì§ˆë¬¸(question)ì— ëŒ€í•œ ë‹µì„ ì¶”ì¶œí•˜ëŠ” ì‚¬ì „í•™ìŠµëœ ëª¨ë¸ ì‚¬ìš©
question_answerer = pipeline("question-answering")

# ì§ˆë¬¸ê³¼ ë¬¸ë§¥ì„ ì…ë ¥í•˜ì—¬ ì •ë‹µì„ ì¶”ì¶œ
result = question_answerer(
    question="Where do I work ?",  # ì§ˆë¬¸
    context="My name is Sylvian and I work at Hugging Face in Brooklyn"  # ë¬¸ë§¥
)

print(result)
```

### ì‚¬ì „ í•™ìŠµëœ ëª¨ë¸ì„ ì´ìš©í•œ ì±—ë´‡

* `model`ì€ `AutoModelForSeq2SeqLM` í´ë˜ìŠ¤ì˜ ì¸ìŠ¤í„´ìŠ¤ì…ë‹ˆë‹¤. ì´ í´ë˜ìŠ¤ëŠ” ì„ íƒí•œ ì–¸ì–´ ëª¨ë¸ê³¼ ìƒí˜¸ì‘ìš©í•  ìˆ˜ ìˆë„ë¡ í•´ì¤ë‹ˆë‹¤.
* `tokenizer`ëŠ” `AutoTokenizer` í´ë˜ìŠ¤ì˜ ì¸ìŠ¤í„´ìŠ¤ì…ë‹ˆë‹¤. ì´ í´ë˜ìŠ¤ëŠ” ì…ë ¥ í…ìŠ¤íŠ¸ë¥¼ ëª¨ë¸ì´ ê°€ì¥ íš¨ìœ¨ì ìœ¼ë¡œ ì´í•´í•  ìˆ˜ ìˆëŠ” ë°©ì‹ìœ¼ë¡œ ë³€í™˜í•´ì¤ë‹ˆë‹¤. êµ¬ì²´ì ìœ¼ë¡œëŠ” í…ìŠ¤íŠ¸ë¥¼ "í† í°(token)"ìœ¼ë¡œ ë¶„í•´í•˜ì—¬ ëª¨ë¸ì´ ì²˜ë¦¬í•  ìˆ˜ ìˆëŠ” í˜•íƒœë¡œ ë°”ê¿”ì¤ë‹ˆë‹¤.

 `"facebook/blenderbot-400M-distill"` ëª¨ë¸ì„ ì‚¬ìš©í•©ë‹ˆë‹¤. ì´ ëª¨ë¸ì€ ì˜¤í”ˆì†ŒìŠ¤ ë¼ì´ì„ ìŠ¤ í•˜ì— ììœ ë¡­ê²Œ ì‚¬ìš©í•  ìˆ˜ ìˆìœ¼ë©°, ì†ë„ë„ ë¹„êµì  ë¹ ë¥´ê¸° ë•Œë¬¸ì— ì„ íƒë˜ì—ˆìŠµë‹ˆë‹¤.
ë‹¤ì–‘í•œ ëª¨ë¸ê³¼ ê·¸ ê¸°ëŠ¥ì„ ë” ì‚´í´ë³´ê³  ì‹¶ë‹¤ë©´ Hugging Face ì›¹ì‚¬ì´íŠ¸ë¥¼ ì°¸ê³ í•´ ë³´ì„¸ìš”: [Hugging Face Models](https://huggingface.co/models).

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# ì‚¬ìš©í•  ëª¨ë¸ ì„ íƒ: "facebook/blenderbot-400M-distill" ì‚¬ìš©
model_name = "facebook/blenderbot-400M-distill"

# ì‚¬ì „í•™ìŠµëœ ì‹œí€€ìŠ¤-íˆ¬-ì‹œí€€ìŠ¤ ì–¸ì–´ ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# í•´ë‹¹ ëª¨ë¸ì— ë§ëŠ” í† í¬ë‚˜ì´ì € ë¶ˆëŸ¬ì˜¤ê¸°
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# ì±—ë´‡ê³¼ ëŒ€í™”í•˜ëŠ” í•¨ìˆ˜ ì •ì˜
def chat_with_bot(tokenizer, model):
    while True:
        # ì‚¬ìš©ìë¡œë¶€í„° ì…ë ¥ ë°›ê¸°
        input_text = input("You: ")

        # ì¢…ë£Œ ì¡°ê±´: ì‚¬ìš©ìê°€ quit, exit, bye ì…ë ¥ ì‹œ ëŒ€í™” ì¢…ë£Œ
        if input_text.lower() in ["quit", "exit", "bye"]:
            print("Chatbot: Goodbye!")
            break

        # ì…ë ¥ ë¬¸ì¥ì„ í† í¬ë‚˜ì´ì¦ˆí•˜ì—¬ í…ì„œë¡œ ë³€í™˜
        inputs = tokenizer.encode(input_text, return_tensors="pt")

        # ëª¨ë¸ë¡œë¶€í„° ì‘ë‹µ ìƒì„± (ìµœëŒ€ 150ê°œì˜ ìƒˆ í† í° ìƒì„±)
        outputs = model.generate(inputs, max_new_tokens=150)

        # ìƒì„±ëœ ì‘ë‹µì„ ë””ì½”ë”©í•˜ì—¬ ì‚¬ëŒì´ ì½ì„ ìˆ˜ ìˆëŠ” ë¬¸ìì—´ë¡œ ë³€í™˜
        response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

        # ì±—ë´‡ì˜ ì‘ë‹µ ì¶œë ¥
        print("Chatbot:", response)

# ëŒ€í™” ì‹œì‘
chat_with_bot(tokenizer, model)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
# ëŒ€í™” ì‹œì‘
chat_with_bot(tokenizer, model)
```
