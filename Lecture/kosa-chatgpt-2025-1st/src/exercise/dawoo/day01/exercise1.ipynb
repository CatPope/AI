```python
import os
import openai
from dotenv import load_dotenv
from pykrx import stock
from pykrx import bond
import matplotlib.pyplot as plt
import pandas as pd
```

```python
load_dotenv()

openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
companys = []
for ticker in stock.get_market_ticker_list():
        종목 = stock.get_market_ticker_name(ticker)
        if "DB" in 종목:
                print(종목, ticker)
                companys.append(종목)
print(companys)
```

```python
company = companys[2]
print(company)
```

```python
df = stock.get_market_ohlcv("20240720", "20240810", "005830")
print(df.head(3))
```

```python
plt.figure(figsize=(15, 3))

plt.plot(df["종가"], color='blue')
plt.plot(df["시가"], color='red')
# plt.plot(df["거래량"], color='green')
plt.legend(("종가", "시가"))
plt.title("DB손해보험 주가")
plt.xlabel("날짜")
plt.ylabel("종가")
```

```python
csv = df.to_csv(index=False)
print(csv, type(csv))
```

```python
model = "gpt-4o-mini"

response = openai.chat.completions.create(
    model=model,
    messages=[
        {
            "role" : "system",
            "content" : "주어진 데이터를 기반으로 해당 주식을 분석하시오."
        },
        {
            "role" : "user",
            "content" : csv
        }
    ],
    max_tokens=1024,
    temperature=0.5,
    top_p=0.8,
    frequency_penalty=0,
    presence_penalty=0,
)

print(response.choices[0].message.content)
```
