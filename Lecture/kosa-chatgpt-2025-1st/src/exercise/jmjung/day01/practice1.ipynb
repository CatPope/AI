```python
# 필요한 라이브러리 임포트
import pandas as pd
from pykrx import stock
import matplotlib.pyplot as plt

# 삼성전자의 종목 코드
ticker = "005930"

# 데이터 가져오기: 최근 1년간의 주가 데이터
start_date = "2023-01-20"
end_date = "2024-01-20"
samsung_data = stock.get_market_ohlcv_by_date(start_date, end_date, ticker)

# 데이터 확인
print(samsung_data.head())

# 주가 데이터 시각화
plt.figure(figsize=(12, 6))
plt.plot(samsung_data.index, samsung_data['종가'], label='종가', color='blue')
plt.title('삼성전자 주가 (종가)')
plt.xlabel('날짜')
plt.ylabel('가격 (원)')
plt.legend()
plt.grid()
plt.show()

# 기본 통계 분석
print("기본 통계:")
print(samsung_data.describe())

# 이동 평균 계산 (20일, 60일)
samsung_data['MA20'] = samsung_data['종가'].rolling(window=20).mean()
samsung_data['MA60'] = samsung_data['종가'].rolling(window=60).mean()

# 이동 평균 시각화
plt.figure(figsize=(12, 6))
plt.plot(samsung_data.index, samsung_data['종가'], label='종가', color='blue')
plt.plot(samsung_data.index, samsung_data['MA20'], label='20일 이동 평균', color='orange')
plt.plot(samsung_data.index, samsung_data['MA60'], label='60일 이동 평균', color='green')
plt.title('삼성전자 주가와 이동 평균')
plt.xlabel('날짜')
plt.ylabel('가격 (원)')
plt.legend()
plt.grid()
plt.show()
```
